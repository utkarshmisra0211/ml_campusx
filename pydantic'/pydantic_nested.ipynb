```python
from pydantic import BaseModel
```

```python
class Address(BaseModel):
    street:str
    city:str
    state:str
    pincode:str
    
```

```python
class Patient(BaseModel):
    name:str
    gender:str
    age:int
    address:Address
    

```

```python
address_info={'street':'123 Main St','city':'Anytown','state':'CA','pincode':'12345'}
address1=Address(**address_info)
print(address1.street)
print(address1.city)
print(address1.state)
print(address1.pincode)
```

```python
patient_info={'name':'Utkarsh','gender':'Male','age':25,'address':address1}
patient1=Patient(**patient_info)
print(patient1.name)
print(patient1.gender)
print(patient1.age)
print(patient1.address.street) 
print(patient1.address.city)
print(patient1.address.state)
print(patient1.address.pincode)
```
